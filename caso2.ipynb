```python
credit_df = credit_df.drop(columns=["V6"])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = credit_df.drop(columns=["Class"])

#etiqueta
y = credit_df["Class"]
```

```python

#X_ent, X_pru, y_ent, y_pru = train_test_split(X, y, test_size=.2)
```

```python

X, y = make_classification(n_samples=1000, n_features=29, n_informative=2, random_state=0, shuffle=False)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.1)

ranForCla = RandomForestClassifier(max_depth=5, random_state=0)

ranForCla.fit(X_train, y_train)

y_pred = ranForCla.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
```
